````python
import numpy as np, pandas as pd
````

````python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00560/SeoulBikeData.csv',encoding= 'unicode_escape')
df
````

````python
# data pre-processing

df=df.drop(['Date', 'Dew point temperature(°C)'], axis=1)

# one-hot encoding for categorical variables
Holiday = pd.get_dummies(df['Holiday'])
Functioning_day = pd.get_dummies(df['Functioning Day'], prefix='function_hour')
Hour = pd.get_dummies(df['Hour'], prefix='hour')
Seasons = pd.get_dummies(df['Seasons'])

df = pd.concat([df.drop(['Holiday', 'Functioning Day', 'Hour', 'Seasons'], axis=1),
                Holiday, Functioning_day, Hour, Seasons], axis=1)

# Normalisation

continuous = ['Temperature(°C)', 'Humidity(%)', 'Wind speed (m/s)', 'Visibility (10m)',
              'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)']

for col in continuous:
    mu= df[col].mean()
    std= df[col].std()
    df[col]-= mu
    df[col]/= std

y = df['Rented Bike Count'] # target variable
X = df.drop(['Rented Bike Count'], axis=1)

X
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
````

**LASSO Regression**   
Ridge Regression    
SVM Regression  
**Random Forest Regression**    
**Neural Network Regression**   

````python
from sklearn.model_selection import RepeatedKFold, GridSearchCV
from sklearn.linear_model import Lasso
````

### LASSO Regression
**Hyperparameters**
1. alpha (Regularisation)
2. max_iter (Maximum number of iterations of optimiser)

````python
model = Lasso()

space = {}
space['alpha'] = [0.05, 0.1, 0.5, 1, 5, 10, 50, 100]
space['max_iter'] = np.arange(1500,2501,50)

cv = RepeatedKFold(n_splits=7, n_repeats=1)

search = GridSearchCV(model, space, cv=cv, scoring='neg_mean_squared_error',n_jobs=-1)

result = search.fit(X_train, y_train)

result.best_params_
````

````python
from sklearn.metrics import mean_squared_error, r2_score
````

````python
model= Lasso(alpha=0.05, max_iter=1500)
model.fit(X_train, y_train)
y_test_pred= model.predict(X_test)

RMSE= np.sqrt(mean_squared_error(y_test, y_test_pred))
r2= r2_score(y_test, y_test_pred)

print("Root Mean Squared Error = %.2f"%RMSE)
print('R2 Score = %.2f'%r2)
````

````python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.neural_network import MLPRegressor
````

### NN Regression
**Hyperparameters**
1. alpha (Regularisation)
2. size of hidden layer

````python
model= MLPRegressor(solver='adam', learning_rate_init=0.05, max_iter=1500)

space={}
space['alpha']= [0.05,0.1,0.5,1,5,10]
space['hidden_layer_sizes']=[(100,),(50,50)]

cv= RepeatedKFold(n_splits=5, n_repeats=1)

search= RandomizedSearchCV(model, space, scoring='neg_mean_squared_error', n_jobs=-1,
                           cv=cv, n_iter=10)

result= search.fit(X_train, y_train)
result.best_params_
````

````python
model = MLPRegressor(alpha=0.05, hidden_layer_sizes=(100,), solver='adam', learning_rate_init=0.05, max_iter=1500)
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)

RMSE = np.sqrt(mean_squared_error(y_test, y_test_pred))
r2 = r2_score(y_test, y_test_pred)

print("Root Mean Squared Error = %.2f"%RMSE)
print('R2 Score = %.2f'%r2)
````

### RF Regression
**Hyperparameters**
1. number of trees
2. depth of tree
3. Minimum samples required to be at a leaf node

````python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

space ={}
space['n_estimators'] = [100,250,500,750]
space['max_depth'] = np.arange(5,11)
space['min_samples_leaf'] = np.arange(1,5)

cv = RepeatedKFold(n_splits=5, n_repeats=1)

search = RandomizedSearchCV(model, space, cv=cv, n_jobs=-1, scoring='neg_mean_squared_error',
                            n_iter=50)

result= search.fit(X_train, y_train)
result.best_params_
````

````python
model = RandomForestRegressor(n_estimators=250 ,max_depth=10 ,max_features='auto' ,min_samples_leaf=1)
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)

RMSE = np.sqrt(mean_squared_error(y_test, y_test_pred))
r2 = r2_score(y_test, y_test_pred)

print("Root Mean Squared Error = %.2f"%RMSE)
print('R2 Score = %.2f'%r2)
````

So, RF and NN regressor are decent models to use. LASSO doesn't predict as accurately as the other two. However, I'm sure that, with more sophisticated models, we can get more accuracy and more satisfication.

````python
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/credit-screening/crx.data', header=None)
df
````

````python
# pre-processing
df[0]= df.replace({0:{'?':np.nan}})[0]
df[1]= pd.to_numeric(df.replace({1:{'?': np.nan}})[1])
df[5]= df.replace({5:{'?':np.nan}})[0]
df[6]= df.replace({6:{'?':np.nan}})[0]
df[13]= pd.to_numeric(df.replace({13:{'?':np.nan}})[13])
df[15]= df.replace({15:{'+':1, '-':0}})[15]
````

````python
df.dtypes
````

````python
df = df.dropna()
df
````

````python
one_hot = pd.get_dummies(df[[0,3,4,5,6,8,9,11,12]])
df = pd.concat([df.drop([0,3,4,5,6,8,9,11,12], axis=1), one_hot], axis=1)

# Normalisation
continuous=[1,2,7,10,13,14]

for col in continuous:
    mu= df[col].mean()
    std= df[col].std()
    df[col]-= mu
    df[col]/= std

df.columns = df.columns.astype(str)
df
````

````python
y = df["15"]
X = df.drop("15", axis=1)
X
````

**NN Classification**   
**Random Forest Classifier**        
SVM 
Regularised Logistic Classifier 

````python
from sklearn.model_selection import train_test_split, RepeatedKFold, RandomizedSearchCV, GridSearchCV
from sklearn.metrics import accuracy_score, f1_score
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier  
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=0)
````

````python
model= MLPClassifier(max_iter=2500, solver='adam',learning_rate_init=0.05)

space={}
space['alpha']= [0.01,0.05,0.1,0.5,1,5,10]
space['hidden_layer_sizes']=[(100,),(50,50),(60,60)]

cv= RepeatedKFold(n_splits=5, n_repeats=1)

search= GridSearchCV(model, space, scoring='f1', n_jobs=-1, cv=cv)

result= search.fit(X_train, y_train)
result.best_params_
````

````python
model= MLPClassifier(alpha=1, hidden_layer_sizes=(100,), max_iter=2500, solver='adam',learning_rate_init=0.05)
model.fit(X_train, y_train)
y_pred= model.predict(X_test)

print('Accuracy = %.2f'%(accuracy_score(y_test, y_pred)))
print('F1 Score is = %.2f'%f1_score(y_test,y_pred))
````

````python
model= RandomForestClassifier()

space={}
space['n_estimators']= [50,100,150,200]
space['max_depth'] = np.arange(6,10)

cv= RepeatedKFold(n_splits=5, n_repeats=1)

search= GridSearchCV(model, space, cv=cv, scoring='f1',n_jobs=-1)

result= search.fit(X_train, y_train)
result.best_params_
````

````python
model=RandomForestClassifier(n_estimators=50, max_depth=9)
model.fit(X_train, y_train)
y_pred= model.predict(X_test)

print('Accuracy = %.2f'%(accuracy_score(y_test, y_pred)))
print('F1 Score is = %.2f'%f1_score(y_test,y_pred))

````

The above models are very usable as indicated by the scores

````python
df0 = pd.read_csv('data/0.csv', header=None)
df1 = pd.read_csv('data/1.csv', header=None)
df2 = pd.read_csv('data/2.csv', header=None)
df3 = pd.read_csv('data/3.csv', header=None)

df = pd.concat([df0,df1,df2,df3])
df
````

````python
y = df[64]
X = df.drop(64, axis=1)

mu = X.mean()
X = X-mu
X
````

````python
from sklearn.decomposition import PCA
m,n = X.shape
components =0

stdRMSE=[]

for i in range(n):
    pcamodel = PCA(n_components=i)
    X_reduced = pcamodel.fit_transform(X)
    X_approx = pcamodel.inverse_transform(X_reduced)

    MSE = np.sum(np.sum((X-X_approx)**2))/m
    var_L2 = (np.var(np.sqrt(np.sum(X**2, axis=1))))
    
    stdRMSE.append(np.sqrt(MSE/var_L2))
    
    if (MSE/var_L2) <= 0.01:
        components = i
        break

components
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)

plt.plot(stdRMSE)
plt.xlabel('Number of Components')
plt.ylabel("Standardised RMSE")
plt.show()
````
